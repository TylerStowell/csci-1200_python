```python
# 1) multiples
# This function prints the first parameter and multiplies it by the repeared number in the for loop. And repeats it to the second
# parameter's value +1.
def multiples(n1, n2):
    y = n2+1
    for x in range(0,y):
        print(n1*x)
```

```python
# 2) star_pyramid
# This function creates a pyramid alined to the right of stars. The parameter inputed is the number or rows and the number of stars in the last 
# row of the pyramid.
def star_pyramid(size):
    for i in range(1,size+1):
        spaces = (size-i) * " "
        stars = (i) * "*"
        print(str(spaces) + str(stars))
```

```python
# 3) star_pyramid_text
# This function creates a pyramid alined to the right of the string in the parameter and the number of times the string
# is printed and the number of rows is the second parameter.
def star_pyramid_text(animal, size):
    for x in range(size):
        animal1 = str(animal)
        spaces = (len(animal1) * (size-x)) * " "
        word = animal1 * (x+1)
        print(str(spaces) + str(word))
```

```python
# 4) factorial
# This function takes the parameter, number, and returns the factoral of the number in the parameter.
def factorial(number):
    count = 1
    for n in range(1,number+1):
        count = count * n
    print(count)
```

```python
# 5) reverse
# This function takes the user's input of a word and returns the word but in reverse order.
def reverse(word):
    words = ""
    for letter in range(len(word)-1, -1, -1):
        words += word[letter]
    print('"' + words + '"')
```

```python
# 6) string_stats
# This function takes the user's input of words sepperated by commas and splits the words by the commas. Then we can count the
# number of vowels, consonances, and spaces in each individual word. We use a for loop to do this so we don't have to write code
# for each word.
def string_stats():
    string = input("Items? ")
    string = string.split(",")

    for word in string:
        a_count = word.count('a')
        e_count = word.count('e')
        i_count = word.count('i')
        o_count = word.count('o')
        u_count = word.count('u')
        vowel_count = a_count + a_count + e_count + i_count + o_count + u_count
        spaces = string.count(" ")
        consonants = len(word) - vowel_count - spaces
        print(" ")
        print(word)
        print("Vowels: " + str(vowel_count))
        print("spaces: " + str(spaces))
        print("consonances: " + str(consonants))
```

```python
# 7) number_square
# This function, takes a paramiter, num, and prints out a square with the number of rows and colums of the square being 
# the value of the paramiter.
def number_square(num):
    for row in range(num+1):
        box = " "
        for columbs in range(num +1):
            box += str(row * columbs)
        print(box)
```

```python
# 8) absolute_day
# this function takes the paramiters for month and day and finds the total day from the 31st of December to the current date
# in the month.
def absolute_day(month,day):
    months = [31,28,30,30,31,29,31,31,30,31,30,31]
    abs_day = 0
    for x in range(0,int(month)-1):
        abs_day += months[x]
    print(abs_day+day)
```
