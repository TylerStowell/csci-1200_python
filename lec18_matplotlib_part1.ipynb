```python
# At the top of your starter notebooks for your data visualization
# projects, there is a cell containing information similar to:

# This line makes it so that your graphs appear in the notebook itself
%matplotlib inline

# This line imports our graphing module and gives it a nickname (plt)
import matplotlib.pyplot as plt

# This line imports the statistics module, which you used in HW 6 and
# may find helpful for your project
import statistics

import random

# Here is an object that gives names to all our columns in our data
class GradeData:
    year = 0
    course_id = 1
    number_students = 2
    average_grade = 3
    
    
def read_data(filename):
    f = open(filename, "r")
    headers = f.readline() # ignore the first row
    print("Here are the headers that we are ignoring: " + str(headers))
    
    rows = [] # list of lists
    for line in f:
        rows.append(line.strip().split(","))
        
    f.close()
    return rows
```
